```python
%matplotlib inline

import matplotlib as mpl
mpl.rcParams['text.color'] = 'w'
mpl.rcParams['axes.labelcolor'] = 'w'
mpl.rcParams['xtick.color'] = 'w'
mpl.rcParams['ytick.color'] = 'w'
mpl.rcParams['xtick.direction'] = 'out'
mpl.rcParams['ytick.direction'] = 'out'
mpl.rcParams['axes.edgecolor'] = 'w'
mpl.rcParams['figure.figsize'] = (15.0, 15.0)

import matplotlib.pyplot as plt
cmap_label = plt.get_cmap('viridis_r')
cmap_label.set_under('#161616')
cmap_bw = plt.get_cmap('viridis')
cmap_bw.set_under('#161616')

import cryptops
import cv2
import numpy as np
import os
import pandas as pd
from skimage import measure
from utils import config, helpers
```

```python
def stitch_4_by_4(lst):
    rows = []
    for r in list(range(4)):
        start = 4*r
        row = np.concatenate((lst[start], lst[start+1], lst[start+2], lst[start+3]), axis=1)
        rows.append(row)
    return np.concatenate((rows[0], rows[1], rows[2], rows[3]), axis=0)
```

```python
# well info
exp = 'ct25_48'
row = 'B'
col = 5

run_type = 'crypt_masks'
flds = list(range(1,17))
bcolor = '#2A323F'
```

```python
# paths
# nuc_im = cv2.imread('test/nuc_objs_stripes.png', cv2.IMREAD_ANYDEPTH)
prefix_template = '/awlab/projects/2015_08_gut_organoid_analysis/processed_data/{exp:s}/merged/combined'
nuc_im_template = 'nuc_objects/{row:s} - {col:02d}(fld {fld:02d} wv UV - DAPI)_nuc_objects.png'
csv_template = 'csv/extracted_csv/{row:s} - {col:02d}(fld {fld:02d} wv UV - DAPI)_nuc_info.csv'
out_path = '/awlab/projects/2015_08_gut_organoid_analysis/analysis_output/{exp:s}/{row:s}{col:02d}_{imtype:s}.png'
```

```python
# read in files
nuc_ims = []
im_datas = []
for fld in flds: 
    prefix = prefix_template.format(exp=exp)
    nuc_im_base = nuc_im_template.format(row=row, col=col, fld=fld)
    csv_base = csv_template.format(row=row, col=col, fld=fld)
    nuc_im_path = os.path.join(prefix, nuc_im_base)
    csv_path = os.path.join(prefix, csv_base)

    nuc_im = cv2.imread(nuc_im_path, cv2.IMREAD_ANYDEPTH)
    im_data = pd.read_csv(csv_path)
    
    nuc_ims.append(nuc_im)
    im_datas.append(im_data)
    
```

```python
# raw DAPI image

fig_dapi = plt.gcf()
plt.imshow(helpers.logical_array(stitch_4_by_4(nuc_ims)), cmap='Greys')
```

```python
# dense nuclei
kn_size = config.get_crypt_finder_dilate_kn_size('crypt_masks')
min_area = config.get_crypt_finder_min_area('crypt_masks')

nuc_denses = []
label_masks = []

for fld in flds:
    nuc_dense, label_mask = cryptops.dense_objects(nuc_ims[fld-1], kernel_size=kn_size, min_area=min_area)
    nuc_denses.append(nuc_dense)
    label_masks.append(label_mask)

fig_dense = plt.gcf()
plt.imshow(stitch_4_by_4(label_masks), cmap=cmap_label, vmin=0.5)
```

```python
# filter EdU
edu_thresh = config.get_crypt_finder_edu_thresh(run_type)
min_edu_num = config.get_crypt_finder_min_edu_num(run_type)

nuc_crypts = []

for fld in flds:
    edu_pos_objs = cryptops.thresh_edu_objects(im_datas[fld-1], edu_thresh)
    nuc_crypt = cryptops.filter_crypts(nuc_denses[fld-1], label_masks[fld-1], edu_pos_objs, min_edu_num)
    nuc_crypts.append(nuc_crypt)

fig_crypts = plt.gcf()
plt.imshow(helpers.logical_array(stitch_4_by_4(nuc_crypts)), cmap=cmap_bw, vmin=0.5)
```

```python
fig_crypts = plt.gcf()
plt.imshow(helpers.logical_array(stitch_4_by_4(nuc_crypts)), cmap='Greys_r')
```

```python
# labeled crypts
nuc_labeleds = []

for fld in flds:
    nuc_labeled = label_masks[fld-1].copy()
    nuc_labeled[nuc_crypts[fld-1]==0] = 0
    nuc_labeleds.append(nuc_labeled)

fig_cryptlabels = plt.gcf()
plt.imshow(stitch_4_by_4(nuc_labeleds), cmap=cmap_label, vmin=0.5)
```

```python
# save images
fig_dapi.savefig(out_path.format(exp=exp, row=row, col=col, imtype='dapi'), facecolor=bcolor,bbox_inches='tight')
fig_dense.savefig(out_path.format(exp=exp, row=row, col=col, imtype='dense'), facecolor=bcolor,bbox_inches='tight')
fig_crypts.savefig(out_path.format(exp=exp, row=row, col=col, imtype='crypts'), facecolor=bcolor,bbox_inches='tight')
fig_cryptlabels.savefig(out_path.format(exp=exp, row=row, col=col, imtype='crypt_labels'), facecolor=bcolor,bbox_inches='tight')
```

```python
import cryptops
exp='ct20a'
run_type = 'merged'
measure_type = 'crypt_measure'
```

```python
# measurements
crypt_measures = ['num_cells', 'num_edu', 'num_paneth', 'nuc_area', 'eccentricity', 'solidity']
well_measures = ['num_cells', 'num_edu', 'num_paneth', 'num_crypt_cells', 'num_crypt_edu', 
'num_crypt_paneth', 'num_villus_cells', 'num_villus_edu', 'num_villus_paneth']

# output directories
crypt_dir = config.get_cp_output_crypt_measure(exp, run_type)
well_dir = config.get_cp_output_well_measure(exp, run_type)

if not os.path.exists(crypt_dir):
os.makedirs(crypt_dir)
if not os.path.exists(well_dir):
os.makedirs(well_dir)

rows = constants.get_96_rows()
cols = constants.get_96_cols()

# matrices of well-level measurements      
well_mats = {k: np.zeros((len(rows), len(cols))) for k in well_measures}
```

```python


for r, row in enumerate(rows):
    for c, col in enumerate(cols):

      # store crypt-level measurements
      crypt_measures = {k: [] for k in crypt_measures}

      # store well-level measurements
      well_nums = {k: [] for k in well_measures}

      # iterate over fields
      num_flds = expinfo.get_num_fields(exp)
      start = expinfo.get_field_start(exp)

      for fld in list(range(start, num_flds+start, 1)):

        im_name = config.build_im_name(exp, row, col, fld, 'dna')
        objs_path = config.get_fpath_crypt_objs(exp, run_type, im_name)
        label_path = config.get_fpath_crypt_labels(exp, run_type, im_name)
        im_data_path = config.get_extract_csv_nuc_outpath(exp, run_type, im_name)

        crypt_objs = cv2.imread(objs_path, cv2.IMREAD_ANYDEPTH)
        label_mask = cv2.imread(label_path, cv2.IMREAD_ANYDEPTH)

        im_data = pd.read_csv(im_data_path)

        edu_thresh = config.get_crypt_finder_edu_thresh(measure_type)
        paneth_thresh = config.get_crypt_finder_paneth_thresh(measure_type)

        edu_objs = thresh_edu_objects(im_data, edu_thresh)
        paneth_objs = thresh_paneth_objects(im_data, paneth_thresh)

        # well-level measurements
        well_measures = {
          'num_cells': len(im_data),
          'num_edu': len(edu_objs),
          'num_paneth': len(paneth_objs),

          'num_crypt_cells': len(nonzero_unique(crypt_objs)),
          'num_crypt_edu': count_stained_objs(crypt_objs, edu_objs),
          'num_crypt_paneth': count_stained_objs(crypt_objs, paneth_objs)}

        well_measures.update({
          'num_villus_cells': well_measures['num_cells'] - well_measures['num_crypt_cells'],
          'num_villus_edu': well_measures['num_edu'] - well_measures['num_crypt_edu'],
          'num_villus_paneth': well_measures['num_paneth'] - well_measures['num_crypt_paneth']
          })


        for k in list(well_nums.keys()):
          well_nums[k].append(well_measures[k])

        # crypt-level measurements
        crypt_measures = measure_crypts_props(crypt_objs, label_mask, edu_objs, paneth_objs, crypt_measures)

      # save well-level measurements
      for k in list(well_mats.keys()):
        well_mats[k][r-1][c-1] = np.mean(well_nums[k])

      # write crypt-level measurements to file
      well_name = config.build_well_name(row, col)
      out_path = config.get_fpath_crypt_measure(exp, run_type, well_name)

      helpers.dict_to_csv(out_path, crypt_measures)
```

```python
output_dir = '/awlab/projects/2015_08_gut_organoid_analysis/test/processed_data/ct20b/merged/combined'
```

```python
str1 = 'Crypt'
str2 = 'Villus'
obj_num_label = 'ObjectNumber'

csv_files = helpers.list_csv_files(output_dir)
csv1_files = [f for f in csv_files if str1 in f]


full_csv1 = csv1_files[0]
# output_csv
```

```python
# create short files for testing
csv1 = full_csv1.replace('.csv', '_short.csv')
full_csv2 = full_csv1.replace(str1, str2)
csv2 = csv1.replace(str1, str2)
merged_csv = csv1.replace(str1, '')
```

```python
# create short files for testing
num_lines = 5
with open(full_csv1, 'r') as f_in:
    freader = csv.reader(f_in, delimiter=',')
    with open(csv1, 'w') as f_out:
        fwriter = csv.writer(f_out, delimiter=',')
        curr_line = 0
        for line in freader:
            if curr_line > num_lines:
                break
            else:
                curr_line += 1
                fwriter.writerow(line)

with open(full_csv2, 'r') as f_in:
    freader = csv.reader(f_in, delimiter=',')
    with open(csv2, 'w') as f_out:
        fwriter = csv.writer(f_out, delimiter=',')
        curr_line = 0
        for line in freader:
            if curr_line > num_lines:
                break
            else:
                curr_line += 1
                fwriter.writerow(line)
```

```python
def check_positives(lst, idx):
    return not [x for i,x in enumerate(lst) if (i in idx) and int(x)<1]
    
def make_negative(lst, idx):
    for i in idx:
        lst[i] = -int(lst[i])
    return lst
def remove_string(str_list, substr):
    return [s.replace(substr, '') for s in str_list]
def index_labels(lst, keyword):
    return [i for i, x in enumerate(lst) if keyword in x]
```

```python
# create merged file
headers = ''
idx = []
with open(merged_csv, 'w') as f_merged:
    fwriter = csv.writer(f_merged, delimiter=',')
    with open(csv1, 'r') as f1:
        freader1 = csv.reader(f1, delimiter=',')
        headers = next(freader1)
        fwriter.writerow(remove_string(headers, str1))
        idx = index_labels(headers, str1)
        idx.append(headers.index(obj_num_label))
        for line in freader1:
            if check_positives(line, idx):
                fwriter.writerow(line)
            else:
                raise ValueError('Negative number detected in {path:s}'.format(path=csv1))
        
    with open(csv2, 'r') as f2:
        freader2 = csv.reader(f2, delimiter=',')
        headers2 = next(freader2)
        headers2 = remove_string(headers, str2)
        if headers2 != headers:
            raise ValueError('File headers needs to match')
        for line in freader2:
            if check_positives(line, idx):
                new_line = make_negative(line, idx)
                fwriter.writerow(new_line)
            else:
                raise ValueError('Negative number detected in {path:s}'.format(path=csv1))
            
```

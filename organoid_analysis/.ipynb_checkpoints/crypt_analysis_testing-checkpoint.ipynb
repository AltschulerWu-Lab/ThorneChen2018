```python
%matplotlib inline

import matplotlib as mpl
mpl.rcParams['text.color'] = 'w'
mpl.rcParams['axes.labelcolor'] = 'w'
mpl.rcParams['xtick.color'] = 'w'
mpl.rcParams['ytick.color'] = 'w'
mpl.rcParams['xtick.direction'] = 'out'
mpl.rcParams['ytick.direction'] = 'out'
mpl.rcParams['axes.edgecolor'] = 'w'
mpl.rcParams['figure.figsize'] = (8.0, 8.0)

import matplotlib.pyplot as plt
cmap_label = plt.get_cmap('viridis_r')
cmap_label.set_under('#161616')
cmap_bw = plt.get_cmap('viridis')
cmap_bw.set_under('#161616')

import cryptops
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from skimage import measure
from utils import config, helpers
```

```python
# well info
exp = 'ct24_control1'
row = 'B'
col = 4
fld = 5

run_type = 'crypt_masks'
bcolor = '#2A323F'
```

```python
# paths
# nuc_im = cv2.imread('test/nuc_objs_stripes.png', cv2.IMREAD_ANYDEPTH)
prefix_template = '/awlab/projects/2015_08_gut_organoid_analysis/processed_data/{exp:s}/merged/combined'
nuc_im_template = 'nuc_objects/{row:s} - {col:02d}(fld {fld:02d} wv UV - DAPI)_nuc_objects.png'
csv_template = 'csv/extracted_csv/{row:s} - {col:02d}(fld {fld:02d} wv UV - DAPI)_nuc_info.csv'
prefix = prefix_template.format(exp=exp)
nuc_im_base = nuc_im_template.format(row=row, col=col, fld=fld)
csv_base = csv_template.format(row=row, col=col, fld=fld)
nuc_im_path = os.path.join(prefix, nuc_im_base)
csv_path = os.path.join(prefix, csv_base)

raw_prefix = '/awlab/projects/2015_08_gut_organoid_analysis/data/{exp:s}'
name_template = '{row:s} - {col:02d}(fld {fld:02d} wv {ch:s})_nuc_info.csv'
paneth_ch = expinfo.get_channel_paneth(exp)
paneth_base = name_template.format(row=row, col=col, fld=fld, ch=paneth_ch)
edu_ch = expinfo.get_channel_edu(exp)
edu_base = name_template.format(row=row, col=col, fld=fld, ch=edu_ch) 
paneth_raw_path = os.path.join(raw_prefix, paneth_base)
edu_raw_path = os.path.join(raw_prefix, edu_base)

out_path = '/awlab/projects/2015_08_gut_organoid_analysis/analysis_output/{exp:s}/{row:s}{col:02d}_fld{fld:02d}_{imtype:s}.png'
```

```python
# read in files
nuc_im = cv2.imread(nuc_im_path, cv2.IMREAD_ANYDEPTH)
im_data = pd.read_csv(csv_path)

# raw DAPI
fig_dapi = plt.gcf()
plt.imshow(helpers.logical_array(nuc_im), cmap='Greys')
```

```python
# dense nuclei
kn_size = config.get_crypt_finder_dilate_kn_size('crypt_masks')
min_area = config.get_crypt_finder_min_area('crypt_masks')
nuc_dense, label_mask = cryptops.dense_objects(nuc_im, kernel_size=kn_size, min_area=min_area)

fig_dense = plt.gcf()
plt.imshow(label_mask, cmap=cmap_label, vmin=0.5)
```

```python
# EdU threshold test

edu_thresh = config.get_crypt_finder_edu_thresh(run_type)
min_edu_num = config.get_crypt_finder_min_edu_num(run_type)

edu_pos_objs = cryptops.thresh_edu_objects(im_data, edu_thresh)
threshed_edu = cryptops.filter_objects(nuc_dense, edu_pos_objs)

# edu_im
fig_edu_thresh = plt.gcf()
plt.imshow(threshed_edu, cmap=cmap_label, vmin=0.5)
```

```python
# filter EdU
edu_thresh = config.get_crypt_finder_edu_thresh(run_type)
min_edu_num = config.get_crypt_finder_min_edu_num(run_type)

edu_pos_objs = cryptops.thresh_edu_objects(im_data, edu_thresh)
nuc_crypt = cryptops.filter_crypts(nuc_dense, label_mask, edu_pos_objs, min_edu_num)

fig_crypts = plt.gcf()
plt.imshow(helpers.logical_array(nuc_crypt), cmap=cmap_bw, vmin=0.5)
```

```python
# labeled crypts
nuc_labeled = label_mask.copy()
nuc_labeled[nuc_crypt==0] = 0

fig_cryptlabels = plt.gcf()
plt.imshow(nuc_labeled, cmap=cmap_label, vmin=0.5)
```

```python
# save images
fig_dapi.savefig(out_path.format(exp=exp, row=row, col=col, fld=fld, imtype='dapi'), bbox_inches='tight')
fig_dense.savefig(out_path.format(exp=exp, row=row, col=col, fld=fld, imtype='dense'), bbox_inches='tight')
fig_crypts.savefig(out_path.format(exp=exp, row=row, col=col, fld=fld, imtype='crypts'), bbox_inches='tight')
fig_cryptlabels.savefig(out_path.format(exp=exp, row=row, col=col, fld=fld, imtype='crypt_labels'), bbox_inches='tight')
```
